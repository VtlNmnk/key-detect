```python
import fileio
import numpy as np
import seaborn as sns
import keys
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X, y = fileio.load_from_range()
```

```python
y_labels = np.argmax(y, axis=0)
```

```python
def number_countplot(ax=None):
    if ax is None:
        ax = plt.gca()
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2.,
                height + 3,
                '{:d}'.format(height),
                ha="center") 
```

```python
g = sns.countplot(y_labels)
number_countplot()
plt.xticks(range(24), [keys.get_string_from_idx(ii).replace('\t', ' ') for ii in range(24)], rotation=90)
plt.show()
```

```python
g = sns.countplot(y_labels < 12)
number_countplot()
plt.xticks(range(2), ['minor', 'major'])
plt.show()
```
